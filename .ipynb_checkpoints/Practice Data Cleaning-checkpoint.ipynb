```python
import numpy as np
import pandas as pd

dummyData = {
    "D1" : [np.nan, "Blue", "Red", "Green", np.nan, "Red", "Green", np.nan, "Blue", "Red"],
    "D2" : ["Green", "Red", "Green", "Blue", "Green", "Blue", np.nan, "Red", "Green", np.nan]
}

colors = pd.DataFrame(dummyData)
colors
```

## Impute missing value with 'Missing' or 'Unknown'

```python
colors.fillna('Missing')
```

## Impute missing value with most common value

```python
print(colors.mode())
colors_imputed = colors.apply(lambda x: x.fillna(x.value_counts().index[0]))
colors_imputed
```
