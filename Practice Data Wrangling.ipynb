```python
import numpy as np
import pandas as pd

dummyData = {
    "Nama" : ['Rivaldi', 'Marco', 'Arnold', 'Desta', 'Mario', 'Asep', 'Maman', 'Ucup', 'Bambang', 'Sukirman'],
    "TinggiBadan" : [185, np.nan, 168, np.nan, 175, 188, 195, 156, np.nan, 148]
}

df = pd.DataFrame(dummyData)
df
```

```python
#Convert type data
df['TinggiObject'] = df['TinggiBadan'].astype(object)
df.info()
df['TinggiFloat'] = df['TinggiObject'].astype(float)
df.info()
```

```python
#imputed missing data
print(df.mean())
df_imputed = df.fillna(df.mean())
df_imputed
```

```python
#Drop NaN

df_drop = df.dropna()
df_drop
```
