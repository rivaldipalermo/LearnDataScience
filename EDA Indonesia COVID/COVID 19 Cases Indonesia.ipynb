```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('cases.csv')
df.head()
```

```python
df.isna()
```

```python
df.info()
```

```python
tanggal = pd.to_datetime(df['date']).dt.strftime('%d')
kasus = df['acc_confirmed']
```

```python
fig, ax = plt.subplots()
plt.plot(tanggal, kasus)
plt.xlabel("Tanggal")
plt.ylabel("Kasus Positif")
fig.autofmt_xdate()
plt.title("COVID 19 Indonesia Bulan Maret 2020")
```

```python
df_prov = pd.read_csv('province_timeline.csv')
df_prov.head()
```

```python
jabar = df_prov['jabar']
jakarta = df_prov['dki_jakarta']
date = df_prov['date']

fig, ax = plt.subplots()
plt.plot(date, jabar)
plt.plot(date, jakarta)
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Positif")
plt.legend(['Jawa Barat', 'DKI Jakarta'])
plt.title('Konfirmasi Kasus Positif COVID-19')
fig.autofmt_xdate()
```
